```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 绘制散点图

```python
# 导入本节示例数据，注意此文件并非本地数据文件，而是从Seaborn库里导入的
# Seaborn库里的所有数据可以在这里查看：https://github.com/mwaskom/seaborn-data
# 如果遇到导入问题，可以从以上GitHub仓库下载到本地，然后读取文件
# 或者是参考https://blog.csdn.net/weixin_51755670/article/details/127979895
df1 = sns.load_dataset("tips")
df1
```

```python
# 传入DataFrame作为数据，以total_bill为横轴数据，tip为纵轴数据，绘制散点图
sns.scatterplot(df1, x="total_bill", y="tip")
plt.show()
```

```python
# 传入Series作为数据，以df1的total_bill为横轴数据，df1的tip为纵轴数据，绘制散点图
sns.scatterplot(x=df1["total_bill"], y=df1["tip"])
plt.show()
```

## 绘制折线图

```python
# 导入示例数据
df2 = sns.load_dataset("flights").query("month == 'Jan'")
df2
```

```python
# 以year为横轴数据，passengers为纵轴数据，绘制折线图
sns.lineplot(df2, x="year", y="passengers")
plt.show()
```

## 绘制条形图

```python
# 导入示例数据
df3 = sns.load_dataset("penguins")
df3
```

```python
# 以species为横轴数据，body_mass_g为纵轴数据，默认聚合操作为求平均值，绘制条形图
sns.barplot(data=df3, x="species", y="body_mass_g")
plt.show()
```

```python
# 以species为横轴数据，body_mass_g为纵轴数据，聚合操作为求最大值，绘制条形图
sns.barplot(data=df3, x="species", y="body_mass_g", 
            estimator=np.min)
plt.show()
```

```python
sns.barplot(data=df3, x="species", y="body_mass_g", 
            estimator=np.max)
plt.show()
```

## 绘制计数图

```python
# 以species为横轴数据，绘制计数图
sns.countplot(data=df3, x="species")
plt.show()
```

## 绘制饼图

```python
# 设置示例数据
df4 = pd.DataFrame({'fruit': ['apple', 'orange', 'banana', 'pear'],
                     'vote': [32, 21, 19, 7]})
df4
```

```python
# 对df4的vote数据绘制饼图
plt.pie(df4['vote'])
plt.show()
```

```python
# 对df4的vote数据绘制饼图，标签对应为df4的fruit数据
plt.pie(df4['vote'], labels=df4['fruit'])
plt.show()
```

```python
# 对df4的vote数据绘制饼图，标签对应为df4的fruit数据，并在图表上展示比例数字，保留1位小数
plt.pie(df4['vote'], labels=df4['fruit'], 
        autopct='%.2f%%')
plt.show()
```

## 更改图表颜色

```python
df1
```

```python
sns.scatterplot(df1[df1['sex'] == 'Female'], x="total_bill", y="tip", color="violet")
plt.show()
```

```python
sns.scatterplot(df1[df1['sex'] == 'Male'], x="total_bill", y="tip", color="green")
plt.show()
```

```python
# 绘制散点图，用 sex 列的值作为分类依据
sns.scatterplot(
    data=df1,
    x="total_bill",
    y="tip",
    hue="sex",  # 根据 sex 列区分颜色
    palette={"Female": "violet", "Male": "green"}  # 指定颜色
)

# 添加图表标题和图例
plt.title("Total Bill vs Tip by Gender")
plt.legend(title="Gender")
plt.show()
```

```python
sns.lineplot(data=df2, x="year", y="passengers",
            color="green")
plt.show()
```

```python
sns.histplot(data=df3, x="body_mass_g",
              color="#c287c7")
plt.show()
```

## 更改图表色盘

```python
sns.set_palette("pastel")
sns.barplot(data=df3, x="species", y="body_mass_g")
plt.show()
```

```python
sns.set_palette("dark")
sns.barplot(data=df3, x="species", y="body_mass_g")
plt.show()
```
